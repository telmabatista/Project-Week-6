```python
import pandas as pd

kick = pd.read_csv('Kickstarter_projects_Feb19.csv')
kick.head()
```

```python
kick.columns
```

```python
kick['year'] = kick['deadline'].astype(str).str[0:4]
kick.year = kick.year.astype(int)
kick
```

```python
kick = kick.drop(['id','name','sub_category','launched_at','deadline','city','state','start_Q','end_Q','currency','country','blurb_length','name_length','year','status'], axis='columns')
kick
```

```python
import pandas as pd

dummies = pd.get_dummies(kick.main_category)
kick = pd.concat([kick,dummies],axis='columns')
kick
```

```python
kick = kick.drop(['main_category','art'], axis='columns')
kick
```

```python
kick.corr()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = kick.drop('usd_pledged',axis='columns')
X
```

```python
y = kick.usd_pledged
y
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
