```python
import pandas as pd


data = pd.read_csv('CAX_Startup_Data.csv')
data.columns
```

```python
data.corr()
```

```python
list1 = data['Dependent-Company Status'].to_list()
new_list = []

for x in list1:
    if x == 'Success':
        new_list.append(1)
    if x == 'Failed':
        new_list.append(0)
        
len(list1)
```

```python
data['label'] = new_list
```

```python
new = data[['label','Internet Activity Score','Last Funding Amount','Number of Co-founders','Number of of advisors','Team size Senior leadership']]
new
```

```python
X = new[['Internet Activity Score','Last Funding Amount','Number of Co-founders','Number of of advisors','Team size Senior leadership']]
y = new.label
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
```
