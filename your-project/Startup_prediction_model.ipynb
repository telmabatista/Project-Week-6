```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('CAX_Startup_Data.csv')
```

```python
data = data[['Dependent-Company Status','Employee Count','Number of Investors in Seed','Number of Investors in Angel and or VC','Number of Co-founders','Number of of advisors','Team size Senior leadership','Number of of repeat investors','Worked in top companies','Have been part of startups in the past?','Have been part of successful startups in the past?','Was he or she partner in Big 5 consulting?','Consulting experience?','Local or global player','B2C or B2B venture?','Average Years of experience for founder and co founder','Breadth of experience across verticals','Years of education','Relevance of education to venture','Relevance of experience to venture','Experience in selling and building products','Dificulty of Obtaining Work force','Pricing Strategy','Time to market service or product','Barriers of entry for the competitors','Company awards','Disruptiveness of technology','Gartner hype cycle stage']]
data.head()
```

```python
data = data.replace('No Info',np.nan)
```

```python
data = data.dropna()
```

```python
data['Employee Count'] = data['Employee Count'].astype(int)
data['Number of Investors in Seed'] = data['Number of Investors in Seed'].astype(int)
data['Number of Investors in Angel and or VC'] = data['Number of Investors in Angel and or VC'].astype(int)
data['Number of of repeat investors'] = data['Number of of repeat investors'].astype(int)
data['Number of Investors in Seed'] = data['Number of Investors in Seed'].astype(int)
data['Local or global player'] = data['Local or global player'].str.lower()
data
```

```python
cat_vars=[
       'Worked in top companies', 'Have been part of startups in the past?',
       'Have been part of successful startups in the past?',
       'Was he or she partner in Big 5 consulting?', 'Consulting experience?',
       'Local or global player', 'B2C or B2B venture?',
       'Average Years of experience for founder and co founder',
       'Breadth of experience across verticals', 'Years of education',
       'Relevance of education to venture',
       'Relevance of experience to venture',
       'Experience in selling and building products',
       'Dificulty of Obtaining Work force', 'Pricing Strategy',
       'Time to market service or product',
       'Barriers of entry for the competitors', 'Company awards',
       'Disruptiveness of technology', 'Gartner hype cycle stage']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(data[var], prefix=var)
    data1=data.join(cat_list)
    data=data1
    
    
cat_vars=['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome']
data_vars=data.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]
```

```python
data_final=data[to_keep]
```

```python
list1 = data['Dependent-Company Status'].to_list()
new_list = []

for x in list1:
    if x == 'Success':
        new_list.append(1)
    else:
        new_list.append(0)
        
len(list1)
```

```python
data_final['status'] = new_list
data_final
```

```python
data_final.to_csv('Clean.csv',index=False)
```

```python
data_final = data_final.drop(['Dependent-Company Status',
       'Worked in top companies', 'Have been part of startups in the past?',
       'Have been part of successful startups in the past?',
       'Was he or she partner in Big 5 consulting?', 'Consulting experience?',
       'Local or global player', 'B2C or B2B venture?',
       'Average Years of experience for founder and co founder',
       'Breadth of experience across verticals', 'Years of education',
       'Relevance of education to venture',
       'Relevance of experience to venture',
       'Experience in selling and building products',
       'Dificulty of Obtaining Work force', 'Pricing Strategy',
       'Time to market service or product',
       'Barriers of entry for the competitors', 'Company awards',
       'Disruptiveness of technology', 'Gartner hype cycle stage'],axis='columns')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

X = data_final.drop(['status'], axis='columns')
y = data_final['status']
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=10000)
logreg.fit(X,y)
y_pred=logreg.predict(X_test)
y_pred=np.round(y_pred)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
